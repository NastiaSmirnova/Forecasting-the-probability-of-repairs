```python
from datetime import datetime
import time
```

```python
def find_nearest(value, array):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
start_time = datetime.now()
df = pd.DataFrame(columns=['carNum', 'operDate', 'operSt','operDestSt','operCargo','operWeight','real_dist','passed_dist_list','passed_dist','was_in_rep'])
# unknown_distance = pd.DataFrame(columns=['operSt', 'operDestSt','Distance'])
sort_aug_copy=sort_aug.copy(deep=True)
for x in range(15600,16000):
    all_movement_for_carriage = sort_aug_copy[(sort_aug_copy['carNum']==all_carNum[x])]
    sort_aug_copy=sort_aug_copy.drop(all_movement_for_carriage.index)
    all_movement_for_carriage = all_movement_for_carriage.reset_index(drop=True)
    print ('Это вагон номер:', x)
    i=0
    passed_dist_list=[]

    flag_rep=0
    k=0
    #условия по формировани. маршрута
    #неизменность веса
    #неизменность станции куда едем
    while i<len(all_movement_for_carriage)-1:
        carr_i=all_movement_for_carriage.iloc[i]
        carr_i1=all_movement_for_carriage.iloc[i+1]
        while (i<len(all_movement_for_carriage)-1) and (carr_i['operWeight']==carr_i1['operWeight'] and
             carr_i['operDestSt']==carr_i1['operDestSt']):
            carr_i=all_movement_for_carriage.iloc[i]
            carr_i1=all_movement_for_carriage.iloc[i+1]

            dist_from_i_to_end = distances[(distances['operSt']==carr_i['operSt'])&(distances['operDestSt']==carr_i['operDestSt'])]['Distance'].values
            dist_from_i1_to_end = distances[(distances['operSt']==carr_i1['operSt'])&(distances['operDestSt']==carr_i1['operDestSt'])]['Distance'].values
#             dist_from_i_to_i1 = distances[(distances['operSt']==carr_i['operSt'])&(distances['operDestSt']==carr_i1['operSt'])]['Distance'].values
            
            if carr_i['operCode']=='РМНТ ':#проверка что ремонт именно текущий
                if len(current_rep[(current_rep['carNum']==carr_i['carNum'])&(current_rep['operDate']==carr_i['operDate'])]):
                    flag_rep=1
                    
            #расчет пройденного расстояния
            if (len(dist_from_i_to_end)>0 and len(dist_from_i1_to_end)>0) :
                passed_dist_list.append(abs(dist_from_i_to_end[0]-dist_from_i1_to_end[0]))

            elif (len(dist_from_i_to_end)==0 and len(dist_from_i1_to_end)>0):
                all_operSt_toDestSt = distances[distances['operDestSt']==carr_i['operDestSt']]
                near_operSt=find_nearest(carr_i['operSt'],all_operSt_toDestSt['operSt'].values)
                dist_from_i_to_end = all_operSt_toDestSt[all_operSt_toDestSt['operSt']==near_operSt]['Distance'].values
#                 unknown_distance = unknown_distance.append({'operSt':carr_i['operSt'], 'operDestSt':carr_i['operDestSt'],'Distance':None}, ignore_index=True)
                passed_dist_list.append(abs(dist_from_i_to_end[0]-dist_from_i1_to_end[0]))
    
            elif (len(dist_from_i1_to_end)==0 and len(dist_from_i_to_end)>0):
                all_operSt_toDestSt = distances[distances['operDestSt']==carr_i1['operDestSt']]
                near_operSt=find_nearest(carr_i1['operSt'],all_operSt_toDestSt['operSt'].values)
                dist_from_i1_to_end = all_operSt_toDestSt[all_operSt_toDestSt['operSt']==near_operSt]['Distance'].values
#                 unknown_distance = unknown_distance.append({'operSt':carr_i1['operSt'], 'operDestSt':carr_i1['operDestSt'],'Distance':None}, ignore_index=True)
                passed_dist_list.append(abs(dist_from_i_to_end[0]-dist_from_i1_to_end[0]))
            
            elif (len(dist_from_i1_to_end)==0 and len(dist_from_i_to_end)==0):
                all_operSt_toDestSt = distances[distances['operDestSt']==carr_i['operDestSt']]
                near_i_operSt=find_nearest(carr_i['operSt'],all_operSt_toDestSt['operSt'].values)
                dist_from_i_to_end = all_operSt_toDestSt[all_operSt_toDestSt['operSt']==near_i_operSt]['Distance'].values
                near_i1_operSt=find_nearest(carr_i1['operSt'],all_operSt_toDestSt['operSt'].values)
                dist_from_i1_to_end = all_operSt_toDestSt[all_operSt_toDestSt['operSt']==near_i1_operSt]['Distance'].values
                passed_dist_list.append(abs(dist_from_i_to_end[0]-dist_from_i1_to_end[0]))
            
            else:
                passed_dist_list.append(None)
                

            i+=1
            print('Это итерация:',i)
        print('выход')
        carr_k=all_movement_for_carriage.iloc[k]
        real_dist=distances[(distances['operSt']==carr_k['operSt'])&(distances['operDestSt']==carr_k['operDestSt'])]['Distance'].values
        if len(real_dist)>0:
            real_dist=real_dist[0]
        else:
            real_dist=None
        if all_movement_for_carriage.iloc[i]['operCode']=='РМНТ ':
            flag_rep=1    
        new_line = {'carNum':carr_k['carNum'],'operDate':carr_k['operDate'],
                'operSt':carr_k['operSt'],'operDestSt':carr_k['operDestSt'],
                'operCargo':carr_k['operCargo'],'operWeight':carr_k['operWeight'],
                'real_dist':real_dist,
                'passed_dist_list': passed_dist_list,
                'passed_dist':sum(np.array(passed_dist_list)[np.array(passed_dist_list) != np.array(None)]),
                'was_in_rep':flag_rep}
        df = df.append(new_line, ignore_index=True)
        flag_rep=0
        passed_dist_list=[]
        k=i+1
        i=i+1
        

print(datetime.now() - start_time)
```
