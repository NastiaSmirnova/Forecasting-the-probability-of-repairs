```python
#определение последних замеров и даты перед маршрутом
uniq_car=routes['carNum'].unique()
df_roots_flanges = pd.DataFrame(columns=['carNum', 'operDate', 'msr_date', 'axle1_rf', 'axle1_lf', 'axle2_rf', 'axle2_lf',
       'axle3_rf', 'axle3_lf', 'axle4_rf', 'axle4_lf'])
for i in range(len(uniq_car)):
    routes_for_i_carr=routes[routes['carNum']==uniq_car[i]].reset_index(drop=True)
    flanges_for_i_carr=flanges[flanges['carNum']==uniq_car[i]].sort_values(by=['msr_date'])

    for j in range(len(routes_for_i_carr)): 
        if len(flanges_for_i_carr)>0:
            st=flanges_for_i_carr[flanges_for_i_carr['msr_date']<=routes_for_i_carr.iloc[j]['operDate']].iloc[-1]
            new_line=pd.concat([st, routes_for_i_carr.iloc[j][['operDate']]])
            df_roots_flanges=df_roots_flanges.append(new_line,ignore_index=True)
```

```python
#число дней после постройки вагона до даты маршрута
uniq_car=routes['carNum'].unique()
df_roots_days_in_operation = pd.DataFrame(columns=['carNum', 'operDate', 'days_in_operation'])
for i in range(len(uniq_car)):
    routes_for_i_carr=routes[routes['carNum']==uniq_car[i]].reset_index(drop=True)

    for j in range(len(routes_for_i_carr)): 
        car = cars[cars['carNum']==uniq_car[i]]
        if len(car)>0:
            days=(routes_for_i_carr['operDate'].iloc[j] - datetime.strptime(car['carManufactureDate'].values[0], "%Y-%m-%d %H:%M:%S.%f")).days
            new_line={'carNum':uniq_car[i],'operDate':routes_for_i_carr['operDate'].iloc[j],'days_in_operation':days}
            df_roots_days_in_operation=df_roots_days_in_operation.append(new_line,ignore_index=True)
```

```python
new2=new.merge(df_roots_days_in_operation,on=['carNum','operDate'])
new2['operDate']=pd.to_datetime(new2['operDate'], format='%Y-%m-%dT')
new2
```

```python
##Число раз когда части вагона ремонтировались на момент маршрута
uniq_car=routes['carNum'].unique()
df_roots_parts = pd.DataFrame(columns=['carNum', 'operDate', 'n_times_parts_in_repair'])
for i in range(len(uniq_car)):
    routes_for_i_carr=routes[routes['carNum']==uniq_car[i]].reset_index(drop=True)
    part = parts[(parts['carNum']==uniq_car[i])]
    for j in range(len(routes_for_i_carr)): 
        p = part[part['repDate']<=routes_for_i_carr['operDate'][j]]
        new_line={'carNum':uniq_car[i],'operDate':routes_for_i_carr['operDate'].iloc[j],'n_times_parts_in_repair':len(p)}
        df_roots_parts=df_roots_parts.append(new_line,ignore_index=True)
```
